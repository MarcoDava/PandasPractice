```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"key":list("ABC")*2,"data1":range(6),"data2":np.arange(5,11)})
```

```python
df
```

```python
group=df.groupby("key")
```

```python
group.agg(["min",np.median,"max"])
```

```python
group.agg({"data1":"min","data2":"max"})
```

```python
def f(x): return x.max()-x.min()
```

```python
group.agg(f)
```
